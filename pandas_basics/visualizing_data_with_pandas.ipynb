```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randn(100, 4)
cols = ['A', 'B', 'C', 'D']

# creating dataframe (df) by seperatly passing data and column names
# we can also create dataframe using python dictionary
df = pd.DataFrame(data = data, columns=cols)
```

```python
df['categorical'] = [1] * 10 + [2] * 20 + [3] * 20 + [4] * 30 + [5] * 20
```

```python
df.head()
```

```python
# checking the distribution of the data with histogram
df['A'].plot.hist()
```

```python
# Now if you want to see the edges of these bins
df['A'].plot.hist(edgecolor = 'k') # here k means black
```

```python
# Now the problem with above is, we can see some empty space along x and y axis we want to remove them
df['A'].plot.hist(edgecolor = 'k').autoscale(enable = True, axis = 'both', tight = True)

# Now there is no empty space along both axis
```

```python
# By default the number of bins is 10, you can increase the bins which will lead to more finer distribution 
df['A'].plot.hist(edgecolor = 'k', bins = 20)

# the problem with this we may see some empty spaces like this, so we may want to decrease bins from 20 to 15
```

```python
# inorder to see the grid 
df['A'].plot.hist(edgecolor = 'k', bins = 20, grid = True)
```

```python
df['categorical'].value_counts()
```

```python
# Now plotting bar graph for categorical variable
df['categorical'].plot.bar(figsize=(14,8))
```

```python
# for horizontal bar plot use barh
df['categorical'].plot.barh(figsize = (12, 8))
```

```python
# now plotting the line plot for column A
df['A'].plot.line(figsize = (12, 7), legend = True)
```

```python
# for entire dataset
df.plot.line(figsize=(12, 7), legend = True)
```

```python
# for greater line width set
df['A'].plot.line(figsize = (12, 7), legend = True, lw=4)
```

```python
# now area plot which are same as line plot which fills the ares
df.plot.area(figsize=(12, 8), legend = True, stacked = False);

# for stacked= True, all the values must be positive or negative
```

```python
# Now scatter plot
df.plot.scatter(x = 'A', y = 'B');
```

```python
# now adding the third dimension using color as column C
df.plot.scatter(x = 'A', y = 'B', c= 'C');
```

```python
# great color map
df.plot.scatter(x = 'A', y = 'B', c= 'C', cmap = 'coolwarm');
```

```python
# now adding the third dimension using size instead of color, but it requires the series instead of just string :|
df.plot.scatter(x = 'A', y = 'B', s= df['C']);

# the warning is because, C contains some negative values and negative size is meaningless
```

```python
# visualizing with bigger size
df.plot.scatter(x = 'A', y = 'B', s= df['C'] * 50);
```

```python
# now also adding the fourth dimension by adding color
df.plot.scatter(x = 'A', y = 'B', s= df['C'] * 50, c='D', cmap='coolwarm');
```

```python
# Now boxplots

# the upper whisker is 25th quartile, middle box is from 25th to 75th quartile and lower whisker is from 75th quartile
# green line is median
df.plot.box(figsize=(12, 8));
```

```python
# now kernel density estimation plot, which is basically estimate of the feature
df.plot.kde();
```

```python
# from this plot we cannot understand, which area has more number of points  
df.plot.scatter(x = 'A', y = 'B')
```

```python
# therefore hexbin is plot which creates hexagonal bins and darker the color of hexagon more the number of points
df.plot.hexbin(x = 'A', y = 'B', gridsize = 25);
```
